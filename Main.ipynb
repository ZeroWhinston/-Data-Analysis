```python
import pandas as pd
import numpy as np
from sklearn import metrics
from functions import TransformColumn, DropColumn, MinMaxFunc, cleanDF, NormalizedDF
from functionsСlassification import KmeansFunc, OneHotFunc
```

```python
#Имена полей
names = ['Partner','EI','CoefSR','AreaOfActivity','CoopTime','CalculationObject','OverduePayment','SalesDebt','DelayDays','ClientDebt']
group_column_name = ['Partner','EI','Coef-SR','AreaOfActivity','CoopTime','DocCount','DocAvg','DocDelayAvg','DocDelayCount','DocRatioOfDelayed']
#Загрузка полей
dataf = pd.read_excel('1.xlsx', names=names)
dataftest = pd.read_excel('2.xlsx', names=names)
```

```python
#Удаление и предобразование обучаего набора данных
df = cleanDF(dataf)
dfNorm = NormalizedDF(df)
dfNorm = dfNorm.round(decimals = 4)
dfKmean = KmeansFunc(dfNorm)
dfCategory = dfKmean[['Partner', 'CategoryOfPartner']]
dfCategory = dfCategory.groupby(by=['Partner'], as_index=False)['CategoryOfPartner'].max()
dfFinal = OneHotFunc(dfKmean)
dfFinal = dfFinal.join(dfKmean[['EI','CoefSR','CoopTime','SalesDebt','DocDelayAvgNorm','DocRatioOfDelayed','DocDelay']])
```

```python
dfFinal
```

```python
#Удаление и предобразование тестового набора данных
testdf = cleanDF(dataftest)
testdfNorm = NormalizedDF(testdf)
testdfNorm = testdfNorm.round(decimals = 4)
testdfNorm['CategoryOfPartner'] = 0
testdfCategory = testdfNorm[['Partner', 'CategoryOfPartner']]
for i in range(testdfCategory.shape[0]):
    df2 = dfCategory
    value = ''
    df3 = ''
    value = testdfCategory.at[i,'Partner']
    df3 = df2.loc[df2['Partner'] == value]
    if df3.size != 0:
        testdfCategory.loc[i,'CategoryOfPartner'] = df3['CategoryOfPartner'].values[0]
    else:
        testdfCategory = testdfCategory.drop(index = i)
        testdfNorm = testdfNorm.drop(index = i)
testdfCategory.reset_index(drop=True, inplace=True)
testdfNorm.reset_index(drop=True, inplace=True)
testdfFinal = OneHotFunc(testdfCategory)
testdfFinal = testdfFinal.join(testdfNorm[['EI','CoefSR','CoopTime','SalesDebt','DocDelayAvgNorm','DocRatioOfDelayed','DocDelay']])
```

```python
#Запись тестового набора в файл
dftest = testdfCategory.join(testdfNorm[['EI','CoefSR','CoopTime','SalesDebt','DocDelayAvgNorm','DocRatioOfDelayed','DocDelay']])
writer = pd.ExcelWriter('CleanTest2.xlsx')
dftest.to_excel(writer)
writer.save()
writer = pd.ExcelWriter('CleanTest.xlsx')
testdfFinal.to_excel(writer)
writer.save()
```

```python
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
```

```python
X = pd.DataFrame(dfFinal, columns = dfFinal.columns[:-1])
y = pd.DataFrame(dfFinal, columns = dfFinal.columns[-1:])
Xt = pd.DataFrame(testdfFinal, columns = testdfFinal.columns[:-1])
yt = pd.DataFrame(testdfFinal, columns = testdfFinal.columns[-1:])
```

```python
tree = DecisionTreeClassifier(criterion='entropy', max_depth = 5).fit(X,y)
print(tree.score(Xt,yt))
print(classification_report(yt, tree.predict(Xt)))
print(confusion_matrix(yt, tree.predict(Xt)))
```

```python
model = LogisticRegression(solver='liblinear', random_state=0).fit(X,y)
print(model.score(Xt,yt))
print(classification_report(yt, model.predict(Xt)))
print(confusion_matrix(yt, model.predict(Xt)))
```

```python
from sklearn.tree import export_graphviz
export_graphviz(
            tree,
            out_file = "myTreeName.dot",
            feature_names = list(X.columns),
            class_names = ['BAD','OK'],
            filled = True,
            rounded = True)
```
